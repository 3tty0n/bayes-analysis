```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import linalg as la
%matplotlib inline 
```

```python
df = pd.read_csv("../data/data_example1.csv", encoding="Shift-JIS")
```

```python
df
```

```python
sapporo = df[df.都道府県 == 1]
sapporo
```

```python
l = list(map(lambda x: x/500000, sapporo['総人口']))
ll = list(map(lambda x: x/10000, sapporo['金融保険業人口']))
sapporo['kakou_jinkou'] = np.array(l)
sapporo['kakou_kinyu'] = np.array(ll)
sapporo
```

```python
H1 = sum(list(map(lambda x: x ** 2, sapporo['kakou_jinkou'])))
H2 = sum(sapporo['kakou_jinkou'])
H3 = H2
H4 = len(sapporo['kakou_jinkou'])

v1 = sum(map(lambda x, y: x * y, sapporo['kakou_jinkou'], sapporo['kakou_kinyu']))
v2 = sum(sapporo['kakou_kinyu'])

H = np.array([[H1, H2], [H3, H4]])
v = np.array([[v1], [v2]])
```

```python
H
```

```python
v
```

```python
a, b = np.dot(la.inv(H), v)
```

```python
print(a, b)
```

```python
x = np.arange(0, 1, 0.1)
y = x * a + b

fig = plt.figure()

ax = fig.add_subplot(1,1,1)

ax.scatter(sapporo['kakou_jinkou'], sapporo['kakou_kinyu'])
ax.plot(x=sapporo['kakou_jinkou'], y=sapporo['kakou_kinyu'])
ax.plot(x, y, "-r", label='y = 0.51482247 * x + 0.02998281')
ax.legend()
```
